```python
import pandas as pd
import random
```

```python
df = pd.read_csv("plots/small_running_dataset.csv", index_col="item")
```

```python
def correct_labels(val):
    if val == "pos":
        return "C"
    elif val == "neg":
        return "D"
    else:
        return val
```

```python
df = df.applymap(correct_labels)
```

```python
df
```

```python
df.drop(columns=['state'], axis=1).to_csv("plots/cats_and_dogs.csv")
```

```python
def plurality(labels, allowable_labels = ['C', 'D']):
    if len(labels) == 0:
        # with no labels, just pick one of the allowable labels at random
        return random.choice(allowable_labels)
    else:
        freqs = dict()
        for val in labels:
            freqs[val] = freqs.get(val, 0) + 1
        max_freq = max(freqs.values())
        winners = [k for k in freqs if freqs[k] == max_freq]
        ## return one of the winners, at random
        return random.choice(winners)
```

```python
def frequency(labels, of_interest='C'):
    labels = list(labels)
    return labels.count(of_interest) / len(labels)
```

```python
cols = [df.iloc[:, 2:2+i].apply(plurality, axis=1) for i in range(1,9)]
pluralities = pd.concat(cols, axis=1).rename(columns = {i:f'k={i+1}' for i in range(8)})
```

```python
pluralities
```

```python
pluralities.to_csv('plots/pluralities.csv')
```

```python
cols = [df.iloc[:, 2:2+i].apply(frequency, axis=1) for i in range(1,9)]
frequencies = pd.concat(cols, axis=1).rename(columns = {i:f'k={i+1}' for i in range(8)})
```

```python
frequencies
```

```python
frequencies.to_csv('plots/frequencies.csv', float_format="%.2f")
```
